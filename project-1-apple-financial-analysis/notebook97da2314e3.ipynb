```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/kaggle/input/datasets/rish59/financial-statements-of-major-companies2009-2023/Financial Statements.csv')

print(df.shape)
print(df.columns.tolist())
print(df.head())
```

```python
# Посмотрим какие компании есть
print(df['Company '].unique())
print(df['Year'].unique())
```

```python
# Фильтруем только AAPL и сортируем по году
aapl = df[df['Company '] == 'AAPL'].sort_values('Year')

# График тренда выручки
plt.figure(figsize=(12, 5))
plt.plot(aapl['Year'], aapl['Revenue'], marker='o', color='blue', linewidth=2)
plt.title('AAPL — Revenue Trend (2009–2023)')
plt.xlabel('Year')
plt.ylabel('Revenue (in M USD)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 5))

ax.plot(aapl['Year'], aapl['Gross Profit'], marker='o', label='Gross Profit', color='green')
ax.plot(aapl['Year'], aapl['Net Income'], marker='s', label='Net Income', color='red')

ax.set_title('AAPL — Gross Profit vs Net Income (2009–2023)')
ax.set_xlabel('Year')
ax.set_ylabel('USD (Millions)')
ax.legend()
ax.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 5))
plt.bar(aapl['Year'], aapl['Net Profit Margin'], color='steelblue')
plt.title('AAPL — Net Profit Margin % (2009–2023)')
plt.xlabel('Year')
plt.ylabel('Net Profit Margin (%)')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 5))

ax.plot(aapl['Year'], aapl['ROE'], marker='o', label='ROE', color='purple')
ax.plot(aapl['Year'], aapl['ROA'], marker='s', label='ROA', color='orange')

ax.set_title('AAPL — ROE vs ROA (2009–2023)')
ax.set_xlabel('Year')
ax.set_ylabel('%')
ax.legend()
ax.grid(True)
plt.tight_layout()
plt.show()
```
